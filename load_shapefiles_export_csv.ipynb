# Load shapefiles export CSV

##### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import shapefile
```

##### Paths

```python
file_path = 'Z://Open_data_sets//World data//Borders//TM_WORLD_BORDERS-0.3.shp'
export_path = 'export//'
```

##### Load the shapefiles and plot

```python

sf = shapefile.Reader(file_path)
shapes = sf.shapes()
csv_array_shapes = np.empty([1,2]) 

for i in range(0,len(shapes)-1):
    curr_coords = np.asarray(shapes[i].points)
    plt.scatter(curr_coords[:,0], curr_coords[:,1], s=0.1)
    csv_array_shapes = np.row_stack((csv_array_shapes, curr_coords)) 
    
print(np.shape(csv_array_shapes))    
```

##### Save the aray as a csv

```python



if not os.path.exists(export_path):
    os.makedirs(export_path)

np.savetxt(export_path+'shapeCoords.csv',csv_array_shapes, fmt='%1.3f', delimiter=',')#, header='X,Y,Z')

```
